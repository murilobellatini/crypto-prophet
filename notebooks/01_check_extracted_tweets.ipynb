```python
import os
import numpy as np
import pandas as pd
from pathlib import Path
cwd = Path(os.getcwd())
data_path = cwd.parents[0] / 'data/tweets'
tweet_paths = [data_path / f for f in os.listdir(data_path) if f.endswith('json')]
df = pd.DataFrame()
for t in tweet_paths:
    df = df.append(pd.read_json(t))
df['user_screen_name'] = df.user.apply(lambda x: x['screen_name'])    
df.groupby(['user_screen_name']).agg({'created_at': [np.min,np.max,len]})
```
