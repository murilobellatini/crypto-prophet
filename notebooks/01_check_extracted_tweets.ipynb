```python
from src.helpers import load_df_tweets
import numpy as np
df = load_df_tweets()
print(df.shape)
print(f"Tweets from {df['created_at'].min()} to {df['created_at'].max()}")
df\
    .groupby(['user_screen_name'])\
    .agg({'created_at': [np.min,np.max,len]})\
    .sort_values([('created_at','amax')], ascending=False)
```
